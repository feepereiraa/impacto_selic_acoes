# O Impacto da Selic no Preço das Ações

A taxa de juros SELIC desempenha um papel crucial no mercado financeiro brasileiro, influenciando diretamente o preço das ações do Ibovespa. A SELIC é a taxa básica de juros estabelecida pelo Banco Central do Brasil e serve como referência para diversas transações financeiras. Quando ela é elevada, os investimentos em renda fixa se tornam mais atraentes, o que pode desestimular investidores a aplicarem em ações. Nesse cenário, é comum observar uma redução na demanda por ações, o que pode levar à queda dos preços do Ibovespa. Por outro lado, quando a taxa SELIC é reduzida, os investimentos em renda fixa se tornam menos atrativos em comparação com as ações, o que pode impulsionar a demanda por ações do Ibovespa e, consequentemente, elevar seus preços. 

Tomando como exemplo os últimos 10 anos, construí alguns gráficos para entender o impacto da variação da SELIC em algumas das principais ações da bolsa brasileira. Os setores e ações foram escolhidos por uma preferência pessoal, muito influenciada pelo grande investidor Luiz Barsi Filho, que sempre tece elogios as ações BEST (Bancos, Energia, Saneamento e Telecomunicações), no caso optei por trocar as ações de telecomunicações por ações das duas maiores empresas da bolsa em volume de negociações, Petrobrás e Vale.

No primeiro gráfico podemos observar o comportamento da taxa de juros nos últimos anos, onde fica claro três períodos distintos, uma alta entre 2012 e 2016, uma queda entre 2016 e 2021 e por fim uma nova alta até o presente momento.


```python
import matplotlib.pyplot as plt
imagem = plt.imread('selic_historico.png')
plt.figure(figsize=(12, 7))
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

Nos quatro gráficos seguintes (2, 3, 4 e 5) é mostrado o comportamento dos preços das ações dos diferentes setores, onde está marcado com uma linha vermelha os momentos em que se inverteu a curva da taxa de juros SELIC. O que fica evidente durante esses períodos, é que as ações tendem a subir de maneira mais acelerada quando a taxa SELIC está em baixa, e subir pouco ou “andar de lado” quando a ela está em alta. Devo ainda ressaltar a presença da pandemia do Covid 19 no início de 2020, que na época gerou uma desvalorização de quase 50% do índice IBOV.

No Gráfico 6 é mostrado a correlação de Pearson entre a variação da taxa SELIC e a variação do preço das ações. Lembrando que essa correlação varia entre -1 e 1, sendo que em intervalos de 0 a 0,3 (negativos ou positivos) a correlação é considerada fraca, de 0,3 a 0,7 moderada e acima de 0,7 é considerada forte. Mesmo nenhuma das ações apresentando correlação forte com a taxa, o que chama atenção é que todas as ações apresentam correlações negativas, o que significa que possuem relação inversamente proporcionais a taxa. Outro ponto a se observar é que alguns setores, como de Saneamento, apresentam correlações mais fortes com a taxa, enquanto outros, como Energia, apresentam correlações mais fracas.

```python
imagem = plt.imread('correlacao_SELIC.png')
plt.figure(figsize=(12, 7))
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

No sétimo gráfico é mostrado o retorno anual médio das ações analisadas nos três momentos distintos da taxa SELIC. Dentre essas ações, é possível observar que, apesar de algumas despontarem mesmo em momentos de alta da SELIC, não existe um movimento uniforme, sendo que algumas apresentaram 
retornos médios negativos ao longo dos anos de alta. Já no momento de queda da taxa, é possível observar retornos mais consistentes, sendo que nenhuma das analisadas entregou menos de 10% ao ano.

```python
imagem = plt.imread('retorno_medio_anual.png')
plt.figure(figsize=(12, 7))
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

Em conclusão, fica evidente o impacto negativo da alta da taxa de juros SELIC nas ações analisadas. Os gráficos mostram que as ações tendem a subir de forma mais acelerada durante os períodos de queda da SELIC, enquanto têm um desempenho mais limitado durante os períodos de alta. Embora haja variações entre os setores e ações específicas, a análise revela uma correlação negativa geral entre a variação da SELIC e a variação dos preços das ações. Portanto, ao analisar e investir em ações, é importante considerar o ambiente macroeconômico e as mudanças na taxa de juros SELIC, pois esses fatores podem influenciar significativamente os retornos dos investimentos em ações.
